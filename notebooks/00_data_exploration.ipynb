```python
prices_reset = prices.reset_index().melt(id_vars="Date", var_name="Ticker", value_name="Price")

fig = px.line(prices_reset, x="Date", y="Price", color="Ticker",
              hover_data={"Ticker": True, "Price": True, "Date": True},
              title="Adjusted Close Price History")

fig.update_layout(
    hovermode="x unified",  
    legend=dict(
        orientation="h",   
        yanchor="bottom",
        y=-0.3,
        xanchor="center",
        x=0.5
    ),
    width=1200,
    height=600
)
fig.show()
```
